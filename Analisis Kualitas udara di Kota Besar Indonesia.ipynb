```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Contoh data kualitas udara (PM2.5) di berbagai kota
kualitas_udara = {
    'Kota': ['Jakarta', 'Bandung', 'Surabaya', 'Medan', 'Yogyakarta', 'Semarang', 'Palembang', 'Makassar'],
    'PM2.5': [85, 65, 70, 78, 60, 68, 73, 66],
    'PM10': [120, 95, 100, 110, 90, 97, 105, 92],
    'NO2': [45, 32, 40, 38, 30, 33, 36, 34],
    'SO2': [20, 15, 18, 19, 14, 16, 17, 15],
    'CO': [1.2, 0.9, 1.0, 1.1, 0.8, 0.95, 1.05, 0.92],
    'O3': [30, 28, 29, 31, 27, 26, 32, 25]
}
```

```python
df = pd.DataFrame(kualitas_udara)
```

```python
# Set style
sns.set(style="whitegrid")
```

```python
# Visualisasi PM2.5
plt.figure(figsize=(10, 6))
sns.barplot(x='Kota', y='PM2.5', data=df, palette='Reds_r')
plt.title('Kadar PM2.5 di Berbagai Kota')
plt.xticks(rotation=40)
plt.ylabel('PM2.5 (µg/m³)')
plt.tight_layout()
plt.show()
```

```python
# Heatmap korelasi antar polutan
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=['Kota']).corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Polutan')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot PM2.5 vs PM10
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PM2.5', y='PM10', hue='Kota', data=df, palette='viridis', s=100)
plt.title('Hubungan PM2.5 dan PM10')
plt.xlabel('PM2.5 (µg/m³)')
plt.ylabel('PM10 (µg/m³)')
plt.tight_layout()
plt.show()
```

```python
# Boxplot untuk distribusi semua jenis polutan
df_melted = df.melt(id_vars='Kota', var_name='Jenis Polutan', value_name='Kadar')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Jenis Polutan', y='Kadar', data=df_melted, palette='Set2')
plt.title('Distribusi Kadar Polutan di Berbagai Kota')
plt.tight_layout()
plt.show()
```
